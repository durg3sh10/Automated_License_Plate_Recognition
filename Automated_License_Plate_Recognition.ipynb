#### Importing the required libraries

```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

#### Read car image used for license plate detection and convert color from BGR to RGB format.

cv2.imread() method loads our image as a multi-dimensional NumPy array (BGR in OpenCv) and cv2.imshow() displays our image to our screen.


```python
car_image = cv2.imread("Downloads/22.jpg")
plt.imshow(cv2.cvtColor(car_image, cv2.COLOR_BGR2RGB))
plt.show()
```

#### To convert an image from one color space to another we use cv2.cvtColor() method, because easier to detect the license plate in a grayscale image.

```python
# Convert our original image from the BGR color space to gray
gray_car_image = cv2.cvtColor(car_image, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray_car_image, cv2.COLOR_BGR2RGB))
plt.show()
```

### Import Haar Cascade XML file for Russian car plate numbers. 
Next we make use of the detectMultiScale method of the CascadeClassifier to run the detection on the grayscale car image.
The key parameters involved in the detectMultiScale function are scaleFactor and minNeighbors.

```python
number_plate_cascade = cv2.CascadeClassifier("Downloads/haarcascades/haarcascades/haarcascade_russian_plate_number.xml")
numberPlates = number_plate_cascade.detectMultiScale(gray_car_image, 1.1, 10)
numberPlates
```

#### The detected objects are returned as a list of bounding boxes (x, y, w, h) representing the location of each plate the classifier found.
We will draw a rectangle given the vertices of the rectangle are known, using the cv2.rectangle() method in OpenCV and  use cv2.putText() method to draw a text string "Number Plate" on the input color image.

```python
color = (250,0, 250)
for x, y, w, h in numberPlates:
    area = w*h
    print(area)
    cv2.rectangle(car_image, (x,y), (x+w, y+h), (255,255,0), 2)
    cv2.putText(car_image,"Number Plate",(x-5,y-5), cv2.FONT_HERSHEY_COMPLEX_SMALL, 1, color, 1)
```

## Display the detected car license plate image

```python
plt.imshow(cv2.cvtColor(car_image, cv2.COLOR_BGR2RGB))
plt.show()
```
